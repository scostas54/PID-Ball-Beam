# Lagrangian Mechanics
In physics, Lagrangian mechanics is a formulation of classical mechanics founded on the stationary-action principle (also known as the principle of least action).
Lagrangian mechanics uses the energies in the system for deriving the equations of motion of a dynamical system.. The central quantity of Lagrangian mechanics is the **Lagrangian (L)**, a function which summarizes the dynamics of the entire system. Overall, the Lagrangian has units of energy.

$$ L = T - V $$ 
where T and V are the kinetic and potential energy of the system.

The stationary action principle requires that the action functional of the system derived from L must remain at a stationary point (a maximum, minimum, or saddle) throughout the time evolution of the system. This constraint allows the calculation of the equations of motion of the system using Lagrange's equations.

$$ \frac{d}{dt}\left ( \frac{\partial L}{\partial \dot{x}} \right )=\frac{\partial L}{\partial x} $$

# The ball and the beam
The ball rolls on the beam without slipping under the action of the force of gravity.  The beam is tilted from an external torque to control the position of the ball on the beam.



- (m)      mass of the ball
- (R)      radius of the ball
- (d)      lever arm offset
- (g)      gravitational acceleration
- (L)      length of the beam
- (J)      ball's moment of inertia
- (r)      ball position coordinate
- (alpha)  beam angle coordinate
- (theta)  servo gear angle

We first define a set of generalized coordinates which fully describe the system. The generalized coordinates are defined as:

$$ q(t)= \begin{bmatrix}
p(t)= L(t)-r(t)\\ 
\alpha (t)
\end{bmatrix} $$



For the Kinetic energy we have:

(Beam, rotational) $$ \begin{equation} \tag{1} K_{beam} =\frac{1}{2}J_{beam}\dot{\alpha ^{2}} \end{equation} $$ 

(Ball, rotational+linear) $$ \begin{equation} \tag{2} K_{ball} =\frac{1}{2}J_{ball}\dot{\alpha_{ball}}^{2} + \frac{1}{2}m_{ball}v_{ball}^{2} \end{equation} $$
where $\dot{\alpha}$ is the angular velocity of the ball and Vb is the linear velocity of the ball. $\dot{\alpha}$ can be expressed as:

$$ \begin{equation} \tag{3} \dot{\alpha_{ball}} = \frac{\dot{p}}{r_{ball}} \end{equation} $$ 

And for $ v_{ball} $ :

$$ \begin{equation} \tag{4} v_{ball}^{2}= \dot{x}^{2} + \dot{y}^{2} \end{equation}$$

$$ \begin{equation} \tag{5} x = p*cos\alpha \end{equation}$$

$$ \begin{equation} \tag{6} \dot{x}= \dot{p}~cos\alpha - p~\dot{\alpha}~sin\alpha \end{equation}$$

$$ \begin{equation} \tag{7} \dot{x}^{2}= \dot{p}^{2}cos^{2}\alpha - 2~p~\dot{p}~\dot{\alpha}~cos\alpha~sin\alpha + p^{2}\dot{\alpha}^{2}sin^{2}\alpha \end{equation}$$

$$ \begin{equation} \tag{8} y = p*sin\alpha \end{equation}$$

$$ \begin{equation} \tag{9} \dot{x}= \dot{p}~sin\alpha + p~\dot{\alpha}~cos\alpha \end{equation}$$

$$ \begin{equation} \tag{10} \dot{x}^{2}= \dot{p}^{2}sin^{2}\alpha + 2~p~\dot{p}~\dot{\alpha}~cos\alpha~sin\alpha + p^{2}\dot{\alpha}^{2}cos^{2}\alpha \end{equation}$$

Substituting (7) and (10) into (4), and knowing that $cos^{2}\alpha + sin^{2}\alpha = 1$ yields:

$$ \begin{equation} \tag{11} v_{ball}^{2}= \dot{p}^{2} + \dot{p}^{2}\dot{\alpha}^{2} \end{equation}$$

Substituting (3) and (11) into (2), we obtain the expression for the kinetic energy of the ball in terms of the generalized coordinates.

$$T=K_{ball} =\frac{1}{2}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\dot{p}^{2}+\frac{1}{2}mp^{2}\dot{\alpha}^{2} $$

Potential energy of the ball is $V = m~g~p~sin\alpha $; thus:

$$ L_{ball} = \frac{1}{2}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\dot{p}^{2}+\frac{1}{2}(mp^{2}+J_{beam})\dot{\alpha}^{2} - mgpsin\alpha $$

If we now apply $ \frac{d}{dt}\left(\frac{\partial L}{\partial \dot{p}}\right)-\frac{\partial L}{\partial p}= 0 $ we have:

$$\frac{\partial L}{\partial \dot{p}}=\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\dot{p}$$
$$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{p}}\right) = \left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\ddot{p}$$
$$\frac{\partial L}{\partial p} = mp\dot{\alpha}^{2} - mgsin\alpha $$

$$\begin{equation} \tag{Equation of motion}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\ddot{p}+mgsin\alpha-mp\dot{\alpha}^{2}= 0 \end{equation}$$

The second derivative of the input angle $\alpha$ actually affects the second derivative of $r$. However, we will ignore this contribution. 

If we apply linearization $ y=(f(a)+f'(a)(x-a))$ about the beam angle, $\alpha = 0$, gives us the following linear approximation of the system:
$$\begin{equation} \tag{12}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\ddot{p} = -mg\alpha \end{equation}$$

The equation which relates the beam angle to the angle of the gear can be approximated as linear by $ \alpha=\frac{d}{L}\Theta$ and therefore:
$$\begin{equation} \tag{13}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\ddot{p} = -mg\frac{d}{L}\Theta \end{equation}$$

# Transfer function

Laplace table:
|$f(t)$|${L}\{f(t)\}$|
|---|---|
|${f}'(t)$|$sF(s)-f(0)$|
|${f}''(t)$|$s^{2}F(s)-sf(0)-{f}'(0)$|

Taking the Laplace transform of equation 13 we get:
$$\begin{equation} \tag{14}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right)\left[s^{2}P(s)-sp(0)-{p}'(0)\right]= -mg\frac{d}{L}\Theta(s) \end{equation}$$
Asuming that at initial conditions (0) the ball is at zero position --> $sp(0)=0$ and with no velocity --> ${p}'(0)=0$, thus:
$$\begin{equation} \tag{15}\left (\frac{J_{ball}}{r_{ball}^{2}}+m\right) s^{2}P(s)= -mg\frac{d}{L}\Theta(s) \end{equation}$$
Rearranging we find the transfer function from the gear angle ($\Theta(s)$) to the ball position ($R(s)$):
$$\begin{equation} \tag{16} G(s) = \frac{P(s)}{\Theta(s)}=-\frac{mgd}{L\left(\frac{J_{ball}}{r_{ball}^{2}}+m\right)}\frac{1}{s^{2}}\left [\frac{m}{rad}\right] \end{equation}$$
**This is the open-loop transfer function of the plant for the ball and beam experiment.**

```python
import control as ctl
import numpy as np
import pandas as pd
import	matplotlib.pyplot as plt
```

```python
m = 0.0027
R = 0.02
g = -9.8
L = 0.3
d = 0.02
J = (2/3)*(m*R**2) #J = 2/3(mR) for a hollow sphere
s = ctl.tf('s')
print(J)
```

```python
s = ctl.tf('s')
P_ball = -m*g*d/L/(J/R**2+m)/s**2
```

```python
[poles,zeros]=ctl.pzmap(s)
print('Poles',poles)
print('Zeros',zeros)
```

Ball's response to a step input on the motor servo gear angle $\theta$ (1-radian step). From this plot it is clear that the system is unstable in open-loop causing the ball to roll right off the end of the beam. Therefore, some method of controlling the ball's position in this system is required.

```python
(time, output)=ctl.step_response(P_ball)
plt.plot(time, output)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

The design criteria for this problem are:

-   Settling time less than 3 seconds
-   Overshoot less than 5%



The transfer function for a PID controller is:
$$\begin{equation} \tag{17} C(s) = K_{p}+\frac{K_{i}}{s}+K_{d}s=-\frac{K_{p}s+K_{i}+K_{d}s^{2}}{s} \end{equation}$$

```python
Kp = 10
C = Kp
sys_cl=ctl.feedback(C*P_ball,1)
(time, output)=ctl.step_response(0.25*sys_cl)
plt.plot(time, output)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

System is unstable with the addition of a proportional gain. We will add below a derivative term to the controller.

```python
Kp = 10
Kd = 10
Ki=0
C = (Kp*s+Ki+Kd*s**2)/s
t = np.linspace(0, 5, 1000) #Tempo de 0s a 5s
sys_cl=ctl.feedback(C*P_ball,1)
_, y = ctl.step_response(0.25*sys_cl,t) #calculando a resposta 
plt.plot(t,y)
plt.xlabel('Time in seconds')
plt.ylabel('Amplitude')
plt.title('Step response')
plt.grid()
```

The system is stable but the overshoot is much too high and the settling time needs to go down a bit. From the PID tutorial page in the section on characteristics of P, I, and D controllers, we see that by increasing $K_d$ we can lower the overshoot and decrease the settling time slightly.

```python
Kp = 50
Kd = 50
Ki = 0
C = (Kp*s+Ki+Kd*s**2)/s
t = np.linspace(0, 5, 1000) #Tempo de 0s a 5s
sys_cl=ctl.feedback(C*P_ball,1)
_, y = ctl.step_response(0.25*sys_cl,t) #calculando a resposta 
plt.plot(t,y)
plt.xlabel('Time in seconds')
plt.ylabel('Amplitude')
plt.title('Step response')
plt.grid()
```
